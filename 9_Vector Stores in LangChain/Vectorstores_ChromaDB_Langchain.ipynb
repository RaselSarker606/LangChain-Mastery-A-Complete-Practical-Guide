Some Libraries

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
```

```python
from langchain.schema import Document

# Create Langchain documents for Movies

doc1 = Document(
    page_content="A bunch of scientists bring back dinosaurs and mayhem breaks loose",
    metadata={"year": 1993, "rating": 7.7, "genre": "science fiction"},
)

doc2 = Document(
    page_content="Leo DiCaprio gets lost in a dream within a dream within a dream within a ...",
    metadata={"year": 2010, "director": "Christopher Nolan", "rating": 8.2},
)

doc3 = Document(
    page_content="A psychologist / detective gets lost in a series of dreams within dreams within dreams and Inception reused the idea",
    metadata={"year": 2006, "director": "Satoshi Kon", "rating": 8.6},
)

doc4 = Document(
    page_content="A bunch of normal-sized women are supremely wholesome and some men pine after them",
    metadata={"year": 2019, "director": "Greta Gerwig", "rating": 8.3},
)

doc5 = Document(
    page_content="Toys come alive and have a blast doing so",
    metadata={"year": 1995, "genre": "animated"},
)


```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
# HF_Embeddding Used
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vector_store = Chroma(
    embedding_function = embedding_model,
    persist_directory = 'chroma_db',
    collection_name = 'sample'
)
```

Add all Document in vector_store


```python
vector_store.add_documents(docs)
```

View all document

```python
vector_store.get(include= ['embeddings', 'documents', 'metadatas'])
```

Search documents

```python
vector_store.similarity_search(
    query = "which discuss about science fiction",
    k = 2  #k =1,2......
)
```

Search with similarity score

```python
vector_store.similarity_search_with_score(
    query = "which discuss about science fiction",
    k = 3  #k =1,2......
)
```

Meta-data-filtering

```python
vector_store.similarity_search_with_score(
    query="",
    filter = {'director': 'Satoshi Kon'}
)
```

Document Update

```python
updated_doc1 = Document(
    page_content="A bunch of scientists bring back dinosaurs and mayhem breaks loose",
    metadata={"year": 1993, "rating": 7.7, "genre": "science fiction", "director": "Steven Spielberg"},
)

vector_store.update_document(document_id = 'd2de7fc1-afe1-466b-9318-dc943ef46ffb', document=updated_doc1)
```
